```julia
using Plots  
using LinearAlgebra  

# Gera 30 valores de x entre 0 e 1
x = range(0, 1, length = 30)

# Definir os coeficientes do polinômio de grau 5
a1, a2, a3, a4, a5, a6 = 1.0, -2.0, 3.0, -1.5, 0.5, 2.0

# Calcular y baseado no polinômio de grau 5 e adicionar ruído
y = a1*x .+ a2*x.^2 .+ a3*x.^3 .+ a4*x.^4 .+ a5*x.^5 .+ a6 .+ randn(30) * 0.1


```

```julia
# Regressão Polinomial de Grau 0
M0 = ones(30, 1)  # Matriz de 30x1 onde todos os elementos são 1 (termo constante)
c0 = y  # Vetor com os valores de y
βsol0 = M0 \ c0  # Solução (média dos valores y)

# Exibir o coeficiente calculado
println("Coeficiente da regressão de grau 0: ", βsol0[1])

# Plotar os pontos e a regressão de grau 0
scatter(x, y, label="Dados", legend=:topleft, title="Regressão Polinomial de Grau 0")
plot!(x, βsol0[1] .* ones(length(x)), label="Grau 0", lw=2)
```

```julia
# Regressão Polinomial de Grau 1
M1 = hcat(ones(30), x)  # Matriz 30x2: primeira coluna é 1 (termo constante), segunda coluna é x
c1 = y  # Vetor com os valores de y
βsol1 = M1 \ c1  # Solução (coeficientes da regressão linear)

# Exibir os coeficientes calculados
println("Coeficientes da regressão de grau 1: ", βsol1)

# Plotar os pontos e a regressão de grau 1
scatter(x, y, label="Dados", legend=:topleft, title="Regressão Polinomial de Grau 1")
plot!(x, βsol1[1] .+ βsol1[2] .* x, label="Grau 1", lw=2)
```

```julia
# Regressão Polinomial de Grau 3
# Matriz 30x4: primeira coluna é 1 (termo constante), segunda coluna é x, terceira coluna é x^2, quarta coluna é x^3
M3 = hcat(ones(30), x, x.^2, x.^3)  
c3 = y  # Vetor com os valores de y
βsol3 = M3 \ c3  # Solução (coeficientes da regressão cúbica)

# Exibir os coeficientes calculados
println("Coeficientes da regressão de grau 3: ", βsol3)

# Plotar os pontos e a regressão de grau 3
scatter(x, y, label="Dados", legend=:topleft, title="Regressão Polinomial de Grau 3")
plot!(x, βsol3[1] .+ βsol3[2] .* x .+ βsol3[3] .* x.^2 .+ βsol3[4] .* x.^3, label="Grau 3", lw=2)
```
