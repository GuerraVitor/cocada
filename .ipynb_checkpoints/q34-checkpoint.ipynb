```julia
#=  Vitor Hugo Guerra 
    123600066     =#

#utilizando a funçao apresentada no material para calculo do min quadrado
function quadmin(x, y, F)
    n = length(x)    #definindo qtde de dados 
    p = length(F)    #definindo qtde de equaçoes 
    M = zeros(p, p)  #criando matriz vazia que recebera somatorios
    c = zeros(p)     #criando vetor vazio que recebero somatorios
    for j = 1:p
        for k = 1:p
            M[j,k] = sum(F[j](x[i]) * F[k](x[i]) for i = 1:n) #montando a primeira matriz (M)
        end
        c[j] = sum(y[i] * F[j](x[i]) for i = 1:n) #montando vetor c 
    end
    β = M \ c #calculando minimos quadrados 
    h(x) = sum(F[j](x) * β[j] for j = 1:p) #calculando modelo encontrado por minimos quadrados 
    return β, h
end

using Plots
#definindo  e plotando dados conhecidos
x = [1.0, 2.0, 4.0]
y = [600, 1400, 3200]
scatter(x, y, lab="dados")
```

```julia
#plotagem dos dados do novo modelo onde foi acrescentado log aos dois lados da equacao
scatter(x, log.(y .- 500), lab="log.(y .- 500)")
```

```julia
#achando os coeficientes por minimos quadrados para o modelo linear que foi encontrado a partir da adicao dos logs
beta, modelo_log = quadmin(x, log.(y .- 500), [x -> 1.0, x -> x])
#plotando o modelo e os dados com a alteraçao do log
scatter(x, log.(y .- 500), lab="log.(y .- 500)")
plot!(modelo_log, extrema(x)..., c=:red, lw=2, lab="modelo log")
```

```julia
#calculando coeficientes do modelo original, isto eh, antes da transformacao com log
alpha = [exp(beta[1]); beta[2]]
#calculando modelo original a partir dos coeficientes originais encontrados
modelo_original(x) = 500 + alpha[1] * exp(log(3) * alpha[2] * x)
#plotando dados e modelo original
scatter(x, y, lab="dados")
plot!(modelo_original , extrema(x)..., c=:red, lw=2, lab="modelo original")
```

```julia
using Roots
#achando a soluçao pedida para o problema
solucao(x) = modelo_original(x) - 24700
horas = find_zero(solucao, 0)
```
