```julia
using Plots

# Dados e cálculos para o primeiro gráfico
erro = [SQR0, SQR1, SQR2, SQR3, SQR4, SQR5, SQR6, SQR7, SQR8, SQR9, SQR10, SQR11, SQR12, SQR13, SQR14, SQR15, SQR16, SQR17, SQR18, SQR19, SQR20, SQR21, SQR22, SQR23, SQR24, SQR25, SQR26, SQR27, SQR28, SQR29, SQR30]
eixoX = 0:30  # Os graus

# Criando o primeiro gráfico
plot1 = plot(eixoX, erro, xlabel="Grau", ylabel="SQR", yaxis=:log10, title="SQR contrabarra", marker=:circle, label=false, size=(800, 400))

# Função para cálculo de mínimos quadrados
function quadmin(x, y, F)
    n = length(x)    # Definindo quantidade de dados 
    p = length(F)    # Definindo quantidade de equações 
    M = zeros(p, p)  # Criando matriz vazia que receberá somatórios
    c = zeros(p)     # Criando vetor vazio que receberá somatórios
    for j = 1:p
        for k = 1:p
            M[j,k] = sum(F[j](x[i]) * F[k](x[i]) for i = 1:n) # Montando a matriz M
        end
        c[j] = sum(y[i] * F[j](x[i]) for i = 1:n)
    end
    β = M \ c # Calculando mínimos quadrados 
    h(x) = sum(F[j](x) * β[j] for j = 1:p) # Calculando modelo encontrado por mínimos quadrados 
    return β, h
end

n = 26 # Grau máximo da regressão polinomial
SQR = zeros(Float64, n + 1)  # Inicializa o vetor para armazenar SQR, agora com n+1 posições

# Regressão para cada grau de 0 até n (26)
for i = 0:n
    β, modelo = quadmin(x, y, [x -> x^j for j = 0:i])  # Regressão polinomial de grau i
    r = y - modelo.(x)  # Calcula o resíduo
    SQR[i + 1] = (sum(r .^ 2))  # Calcula e armazena a soma dos quadrados dos resíduos
end

# Criando o segundo gráfico
plot2 = plot(0:n, SQR, xlabel="Grau", ylabel="SQR", yaxis=:log10, title="SQR função 'quadmin'", marker=:circle, label=false, size=(800, 400))

# Determinando os limites do eixo y com base nos dados
ymin1, ymax1 = minimum(erro), maximum(erro)
ymin2, ymax2 = minimum(SQR), maximum(SQR)
ymin = min(ymin1, ymin2)
ymax = max(ymax1, ymax2)

# Ajustando limites do eixo y
plot1 = plot1 |> (p -> plot!(p, ylims=(ymin, ymax)))
plot2 = plot2 |> (p -> plot!(p, ylims=(ymin, ymax)))

# Exibindo os gráficos lado a lado
plot(plot1, plot2, layout=(1, 2))
```
